## Challenge description

The goal of this challenge is to create a PowerShell script that builds a Server Profile Template for a VMware VSAN node on Synergy. 

## Fork the Challenge Repo

We leverage GitHub to provide the response templates and collect your response. This means that your reponse will be delivered via a Git Pull Request in this [OneViewChallenge](https://github.com/HPEDevCom/OneViewChallenge) repo. 

But before, **you need to fork** this [OneView Challenge](https://github.com/HPEDevCom/OneViewChallenge.git) repository into your own GitHub account.

![GitFork](Pictures/GitFork.png)

## Clone the repo in your Jupyter environment 

Once you have a copy of the repo in your own GitHub account, **you have to clone** it into this Jupyter environment here are some steps to do this:

1. Copy the OneView Challenge GitHub repo location in **your clipboard**. **Be sure to clone your forked copy, not the original repo or you will not be able to submit your work**

![ClipboardGitLocation](Pictures/ClipboardOfGit.png)

2. From your Jupyter environment, **Select `Terminal`** at the bottom left of the main `Launcher` pane, . If you don't have any Launcher pane, click on the `+` sign in the upper left of the left pane.

![LauncherPane](Pictures/StartTerminal.png)

3. Perform the **clone**

Issue the following commands in your terminal, using Ctrl-V to paste the content of your clipboard: 

```Bash
cd Discover/TheOneViewChallenge
git clone https://github.com/<YourGitHubLocation>/OneViewChallenge.git
```

![GitClone](Pictures/GitClone.png)


## Work on your challenge

Once the content of the OneView Challenge repo is downloaded in your Jupyter account, from the left pane, navigate to the `OneviewChallenge` directory and open the notebook you will find there

It's now time for you to do the real work :-)
